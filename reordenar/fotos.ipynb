```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB3, ResNet50, VGG16, InceptionV3
from tensorflow.keras.optimizers import Adam, RMSprop
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
import os

IMG_SIZE = (500, 500)
BATCH_SIZE = 16
EPOCHS = 50

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.8, 1.2],
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    '/content/drive/MyDrive/datos/entrenamiento',
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=True
)

validation_generator = val_datagen.flow_from_directory(
    '/content/drive/MyDrive/datos/validacion',
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False
)

early_stop = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=8, min_lr=1e-7)
checkpoint = ModelCheckpoint('best_model.h5', monitor='val_accuracy', save_best_only=True, mode='max')

def create_model(model_type='custom'):
    if model_type == 'custom_cnn':
        model = Sequential([
            layers.Conv2D(32, (3,3), activation='relu', input_shape=(500, 500, 3)),
            layers.BatchNormalization(),
            layers.MaxPooling2D(2,2),

            layers.Conv2D(64, (3,3), activation='relu'),
            layers.BatchNormalization(),
            layers.MaxPooling2D(2,2),

            layers.Conv2D(128, (3,3), activation='relu'),
            layers.BatchNormalization(),
            layers.MaxPooling2D(2,2),

            layers.Conv2D(256, (3,3), activation='relu'),
            layers.BatchNormalization(),
            layers.MaxPooling2D(2,2),

            layers.GlobalAveragePooling2D(),
            layers.Dense(512, activation='relu'),
            layers.Dropout(0.5),
            layers.Dense(1, activation='sigmoid')
        ])

    elif model_type == 'efficientnet':
        base_model = EfficientNetB3(weights='imagenet', include_top=False, input_shape=(500, 500, 3))
        base_model.trainable = True

        # Congelar las primeras capas y fine-tune las últimas
        for layer in base_model.layers[:-20]:
            layer.trainable = False

        model = Sequential([
            base_model,
            layers.GlobalAveragePooling2D(),
            layers.Dense(256, activation='relu'),
            layers.BatchNormalization(),
            layers.Dropout(0.5),
            layers.Dense(1, activation='sigmoid')
        ])

    elif model_type == 'resnet':
        base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(500, 500, 3))
        base_model.trainable = True

        # Congelar las primeras capas
        for layer in base_model.layers[:-30]:
            layer.trainable = False

        model = Sequential([
            base_model,
            layers.GlobalAveragePooling2D(),
            layers.Dense(256, activation='relu'),
            layers.BatchNormalization(),
            layers.Dropout(0.5),
            layers.Dense(1, activation='sigmoid')
        ])

    elif model_type == 'vgg16':
        base_model = VGG16(weights='imagenet', include_top=False, input_shape=(500, 500, 3))
        base_model.trainable = True

        # Congelar las primeras capas
        for layer in base_model.layers[:-10]:
            layer.trainable = False

        model = Sequential([
            base_model,
            layers.GlobalAveragePooling2D(),
            layers.Dense(256, activation='relu'),
            layers.BatchNormalization(),
            layers.Dropout(0.5),
            layers.Dense(1, activation='sigmoid')
        ])

    elif model_type == 'inception':
        base_model = InceptionV3(weights='imagenet', include_top=False, input_shape=(500, 500, 3))
        base_model.trainable = True

        # Congelar las primeras capas
        for layer in base_model.layers[:-30]:
            layer.trainable = False

        model = Sequential([
            base_model,
            layers.GlobalAveragePooling2D(),
            layers.Dense(256, activation='relu'),
            layers.BatchNormalization(),
            layers.Dropout(0.5),
            layers.Dense(1, activation='sigmoid')
        ])

    return model

# Diccionario para almacenar historiales
history_dict = {}

# Lista de modelos a probar
models_to_try = ['resnet','efficientnet', 'custom_cnn','inception', 'vgg16']

for model_type in models_to_try:
    print(f"\n{'='*50}")
    print(f"Entrenando modelo: {model_type}")
    print(f"{'='*50}")

    # Crear y compilar modelo
    model = create_model(model_type)

    # Usar diferentes optimizadores según el modelo
    if model_type == 'custom_cnn':
        optimizer = Adam(learning_rate=0.0001)
    else:
        optimizer = Adam(learning_rate=0.00001)

    model.compile(optimizer=optimizer,
                  loss='binary_crossentropy',
                  metrics=['accuracy',
                           tf.keras.metrics.Precision(name='precision'),
                           tf.keras.metrics.Recall(name='recall')])

    # Entrenar modelo
    history = model.fit(
        train_generator,
        epochs=EPOCHS,
        validation_data=validation_generator,
        callbacks=[early_stop, reduce_lr, checkpoint],
        verbose=1
    )

    # Guardar historial
    history_dict[model_type] = history

    # Guardar modelo
    model.save(f'{model_type}_model.h5')

    # Evaluar modelo
    print(f"\nEvaluando modelo {model_type}:")
    evaluation = model.evaluate(validation_generator)
    print(f"Pérdida: {evaluation[0]:.4f}, Precisión: {evaluation[1]:.4f}")

# Graficar resultados comparativos
plt.figure(figsize=(20, 15))

# Precisión de entrenamiento y validación
plt.subplot(2, 3, 1)
for model_type in models_to_try:
    plt.plot(history_dict[model_type].history['accuracy'],
             label=f'{model_type} Train', linestyle='--')
    plt.plot(history_dict[model_type].history['val_accuracy'],
             label=f'{model_type} Val')
plt.title('Precisión de Entrenamiento y Validación')
plt.ylabel('Precisión')
plt.xlabel('Época')
plt.legend()
plt.grid(True)

# Pérdida de entrenamiento y validación
plt.subplot(2, 3, 2)
for model_type in models_to_try:
    plt.plot(history_dict[model_type].history['loss'],
             label=f'{model_type} Train', linestyle='--')
    plt.plot(history_dict[model_type].history['val_loss'],
             label=f'{model_type} Val')
plt.title('Pérdida de Entrenamiento y Validación')
plt.ylabel('Pérdida')
plt.xlabel('Época')
plt.legend()
plt.grid(True)

# Precisión de validación comparativa
plt.subplot(2, 3, 3)
for model_type in models_to_try:
    plt.plot(history_dict[model_type].history['val_accuracy'],
             label=model_type)
plt.title('Comparación de Precisión en Validación')
plt.ylabel('Precisión')
plt.xlabel('Época')
plt.legend()
plt.grid(True)

# Learning rate durante el entrenamiento
plt.subplot(2, 3, 4)
for model_type in models_to_try:
    if 'lr' in history_dict[model_type].history:
        plt.plot(history_dict[model_type].history['lr'],
                 label=model_type)
plt.title('Tasa de Aprendizaje Durante el Entrenamiento')
plt.ylabel('Learning Rate')
plt.xlabel('Época')
plt.legend()
plt.grid(True)
plt.yscale('log')

# Precisión y Recall en validación
plt.subplot(2, 3, 5)
val_metrics = {}
for model_type in models_to_try:
    val_precision = max(history_dict[model_type].history['val_precision'])
    val_recall = max(history_dict[model_type].history['val_recall'])
    val_metrics[model_type] = {'precision': val_precision, 'recall': val_recall}

    plt.scatter(val_precision, val_recall, s=100, label=model_type)
    plt.annotate(model_type, (val_precision, val_recall))

plt.title('Precisión vs Recall en Validación')
plt.xlabel('Precisión')
plt.ylabel('Recall')
plt.grid(True)
plt.legend()

# Resumen de métricas finales
plt.subplot(2, 3, 6)
final_accuracies = [max(history_dict[model_type].history['val_accuracy']) for model_type in models_to_try]
plt.bar(models_to_try, final_accuracies)
plt.title('Precisión Final en Validación por Modelo')
plt.ylabel('Precisión')
plt.xticks(rotation=45)
plt.grid(True)

plt.tight_layout()
plt.savefig('comparacion_modelos.png', dpi=300, bbox_inches='tight')
plt.show()

# Identificar el mejor modelo
best_model_name = max(history_dict, key=lambda x: max(history_dict[x].history['val_accuracy']))
print(f"\n{'='*50}")
print(f"Mejor modelo: {best_model_name}")
print(f"{'='*50}")

# Cargar el mejor modelo
best_model = tf.keras.models.load_file(f'{best_model_name}_model.h5')

# Matriz de confusión y reporte de clasificación del mejor modelo
print(f"\nEvaluación detallada del mejor modelo ({best_model_name}):")

# Predecir con el mejor modelo
y_pred = best_model.predict(validation_generator)
y_pred = (y_pred > 0.5).astype(int)
y_true = validation_generator.classes

# Matriz de confusión
plt.figure(figsize=(10, 8))
cm = confusion_matrix(y_true, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=validation_generator.class_indices.keys(),
            yticklabels=validation_generator.class_indices.keys())
plt.title(f'Matriz de Confusión - {best_model_name}')
plt.ylabel('Verdadero')
plt.xlabel('Predicción')
plt.savefig('matriz_confusion.png', dpi=300, bbox_inches='tight')
plt.show()

# Reporte de clasificación
print('\nReporte de Clasificación:')
print(classification_report(y_true, y_pred,
                            target_names=validation_generator.class_indices.keys()))

# Visualizar algunas predicciones
def visualize_predictions(model, generator, num_images=10):
    plt.figure(figsize=(15, 10))
    images, labels = next(generator)
    predictions = model.predict(images)

    for i in range(num_images):
        plt.subplot(2, 5, i+1)
        plt.imshow(images[i])
        true_label = 'Llama' if labels[i] == 0 else 'Alpaca'
        pred_label = 'Llama' if predictions[i] < 0.5 else 'Alpaca'
        confidence = predictions[i][0] if predictions[i] < 0.5 else 1 - predictions[i][0]

        color = 'green' if true_label == pred_label else 'red'
        plt.title(f'True: {true_label}\nPred: {pred_label}\nConf: {confidence:.2f}', color=color)
        plt.axis('off')

    plt.tight_layout()
    plt.savefig('predicciones_ejemplo.png', dpi=300, bbox_inches='tight')
    plt.show()

# Visualizar predicciones con el mejor modelo
visualize_predictions(best_model, validation_generator)

# Resumen de todos los modelos
print(f"\n{'='*60}")
print("RESUMEN COMPARATIVO DE TODOS LOS MODELOS")
print(f"{'='*60}")

for model_type in models_to_try:
    history = history_dict[model_type]
    best_val_acc = max(history.history['val_accuracy'])
    best_val_precision = max(history.history['val_precision'])
    best_val_recall = max(history.history['val_recall'])
    epochs_to_converge = len(history.history['val_accuracy'])

    print(f"\n{model_type}:")
    print(f"  - Mejor precisión validación: {best_val_acc:.4f}")
    print(f"  - Mejor precisión: {best_val_precision:.4f}")
    print(f"  - Mejor recall: {best_val_recall:.4f}")
    print(f"  - Épocas hasta convergencia: {epochs_to_converge}")

# Guardar resultados completos
import json
results = {}

for model_type in models_to_try:
    results[model_type] = {
        'val_accuracy': history_dict[model_type].history['val_accuracy'],
        'val_loss': history_dict[model_type].history['val_loss'],
        'accuracy': history_dict[model_type].history['accuracy'],
        'loss': history_dict[model_type].history['loss'],
        'val_precision': history_dict[model_type].history['val_precision'],
        'val_recall': history_dict[model_type].history['val_recall']
    }

with open('resultados_entrenamiento.json', 'w') as f:
    json.dump(results, f)

print(f"\nResultados guardados en 'resultados_entrenamiento.json'")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import (
    EfficientNetB0, EfficientNetB3, ResNet50, VGG16, InceptionV3,
    DenseNet121, MobileNetV2, Xception, NASNetMobile
)
from tensorflow.keras.optimizers import Adam, RMSprop, SGD, Nadam
from tensorflow.keras.callbacks import (
    EarlyStopping, ReduceLROnPlateau, ModelCheckpoint,
    TensorBoard, CSVLogger, LearningRateScheduler
)
from tensorflow.keras.regularizers import l2
from tensorflow.keras.layers import (
    Conv2D, MaxPooling2D, GlobalAveragePooling2D,
    Dense, Dropout, BatchNormalization, Activation,
    Input, concatenate, SpatialDropout2D
)
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
import os
from datetime import datetime
import json
import pandas as pd
from keras_tuner import RandomSearch, HyperParameters

# Configuración
IMG_SIZE = (500, 500)
BATCH_SIZE = 16
EPOCHS = 100
INIT_LR = 0.0001

# Rutas de datos
train_dir = '/content/drive/MyDrive/datos/entrenamiento'
val_dir = '/content/drive/MyDrive/datos/validacion'

# Verificar distribución de clases
def check_class_distribution(directory):
    classes = os.listdir(directory)
    class_counts = {}
    for cls in classes:
        class_path = os.path.join(directory, cls)
        if os.path.isdir(class_path):
            class_counts[cls] = len(os.listdir(class_path))
    return class_counts

train_class_counts = check_class_distribution(train_dir)
val_class_counts = check_class_distribution(val_dir)

print("Distribución de clases en entrenamiento:", train_class_counts)
print("Distribución de clases en validación:", val_class_counts)

# Calcular pesos de clases para manejo de desbalance
class_weights = compute_class_weight(
    'balanced',
    classes=np.array([0, 1]),
    y=np.concatenate([np.zeros(train_class_counts['llama']), np.ones(train_class_counts['alpaca'])])
)
class_weights = dict(enumerate(class_weights))
print("Pesos de clases:", class_weights)

# Data augmentation mejorado
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=45,
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.3,
    zoom_range=0.3,
    horizontal_flip=True,
    vertical_flip=True,
    brightness_range=[0.7, 1.3],
    contrast_range=[0.8, 1.2],
    channel_shift_range=50.0,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=True,
    seed=42
)

validation_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False
)

# Callbacks mejorados
def create_callbacks(model_name):
    early_stop = EarlyStopping(
        monitor='val_loss',
        patience=20,
        restore_best_weights=True,
        verbose=1,
        mode='min'
    )

    reduce_lr = ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=8,
        min_lr=1e-8,
        verbose=1,
        mode='min'
    )

    checkpoint = ModelCheckpoint(
        f'best_{model_name}.h5',
        monitor='val_accuracy',
        save_best_only=True,
        mode='max',
        verbose=1
    )

    # TensorBoard callback
    log_dir = f"logs/{model_name}/" + datetime.now().strftime("%Y%m%d-%H%M%S")
    tensorboard = TensorBoard(
        log_dir=log_dir,
        histogram_freq=1,
        update_freq='epoch'
    )

    # CSV Logger
    csv_logger = CSVLogger(f'training_{model_name}.csv', append=True)

    return [early_stop, reduce_lr, checkpoint, tensorboard, csv_logger]

# Modelos mejorados
def create_custom_cnn_v2():
    model = Sequential([
        # Bloque 1
        Conv2D(32, (3, 3), activation='relu', input_shape=(500, 500, 3)),
        BatchNormalization(),
        Conv2D(32, (3, 3), activation='relu'),
        BatchNormalization(),
        MaxPooling2D(2, 2),
        SpatialDropout2D(0.1),

        # Bloque 2
        Conv2D(64, (3, 3), activation='relu'),
        BatchNormalization(),
        Conv2D(64, (3, 3), activation='relu'),
        BatchNormalization(),
        MaxPooling2D(2, 2),
        SpatialDropout2D(0.2),

        # Bloque 3
        Conv2D(128, (3, 3), activation='relu'),
        BatchNormalization(),
        Conv2D(128, (3, 3), activation='relu'),
        BatchNormalization(),
        MaxPooling2D(2, 2),
        SpatialDropout2D(0.3),

        # Bloque 4
        Conv2D(256, (3, 3), activation='relu'),
        BatchNormalization(),
        Conv2D(256, (3, 3), activation='relu'),
        BatchNormalization(),
        MaxPooling2D(2, 2),
        SpatialDropout2D(0.4),

        # Clasificación
        GlobalAveragePooling2D(),
        Dense(512, activation='relu', kernel_regularizer=l2(0.001)),
        BatchNormalization(),
        Dropout(0.5),
        Dense(256, activation='relu', kernel_regularizer=l2(0.001)),
        BatchNormalization(),
        Dropout(0.5),
        Dense(1, activation='sigmoid')
    ])
    return model

def create_custom_cnn_v3():
    inputs = Input(shape=(500, 500, 3))

    # Primer camino (más profundo)
    x1 = Conv2D(32, (3, 3), activation='relu')(inputs)
    x1 = BatchNormalization()(x1)
    x1 = Conv2D(32, (3, 3), activation='relu')(x1)
    x1 = BatchNormalization()(x1)
    x1 = MaxPooling2D(2, 2)(x1)
    x1 = SpatialDropout2D(0.1)(x1)

    x1 = Conv2D(64, (3, 3), activation='relu')(x1)
    x1 = BatchNormalization()(x1)
    x1 = Conv2D(64, (3, 3), activation='relu')(x1)
    x1 = BatchNormalization()(x1)
    x1 = MaxPooling2D(2, 2)(x1)
    x1 = SpatialDropout2D(0.2)(x1)

    x1 = Conv2D(128, (3, 3), activation='relu')(x1)
    x1 = BatchNormalization()(x1)
    x1 = Conv2D(128, (3, 3), activation='relu')(x1)
    x1 = BatchNormalization()(x1)
    x1 = MaxPooling2D(2, 2)(x1)
    x1 = SpatialDropout2D(0.3)(x1)

    # Segundo camino (más ancho)
    x2 = Conv2D(64, (5, 5), activation='relu')(inputs)
    x2 = BatchNormalization()(x2)
    x2 = MaxPooling2D(2, 2)(x2)
    x2 = SpatialDropout2D(0.1)(x2)

    x2 = Conv2D(128, (5, 5), activation='relu')(x2)
    x2 = BatchNormalization()(x2)
    x2 = MaxPooling2D(2, 2)(x2)
    x2 = SpatialDropout2D(0.2)(x2)

    x2 = Conv2D(256, (5, 5), activation='relu')(x2)
    x2 = BatchNormalization()(x2)
    x2 = MaxPooling2D(2, 2)(x2)
    x2 = SpatialDropout2D(0.3)(x2)

    # Concatenar caminos
    x = concatenate([x1, x2])

    # Capas finales
    x = Conv2D(512, (3, 3), activation='relu')(x)
    x = BatchNormalization()(x)
    x = GlobalAveragePooling2D()(x)

    x = Dense(512, activation='relu', kernel_regularizer=l2(0.001))(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)

    x = Dense(256, activation='relu', kernel_regularizer=l2(0.001))(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)

    outputs = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=inputs, outputs=outputs)
    return model

def create_model_with_transfer_learning(base_model_name, input_shape=(500, 500, 3)):
    # Seleccionar modelo base
    if base_model_name == 'EfficientNetB0':
        base_model = EfficientNetB0(weights='imagenet', include_top=False, input_shape=input_shape)
    elif base_model_name == 'EfficientNetB3':
        base_model = EfficientNetB3(weights='imagenet', include_top=False, input_shape=input_shape)
    elif base_model_name == 'ResNet50':
        base_model = ResNet50(weights='imagenet', include_top=False, input_shape=input_shape)
    elif base_model_name == 'VGG16':
        base_model = VGG16(weights='imagenet', include_top=False, input_shape=input_shape)
    elif base_model_name == 'InceptionV3':
        base_model = InceptionV3(weights='imagenet', include_top=False, input_shape=input_shape)
    elif base_model_name == 'DenseNet121':
        base_model = DenseNet121(weights='imagenet', include_top=False, input_shape=input_shape)
    elif base_model_name == 'MobileNetV2':
        base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=input_shape)
    elif base_model_name == 'Xception':
        base_model = Xception(weights='imagenet', include_top=False, input_shape=input_shape)
    elif base_model_name == 'NASNetMobile':
        base_model = NASNetMobile(weights='imagenet', include_top=False, input_shape=input_shape)
    else:
        raise ValueError(f"Modelo base no soportado: {base_model_name}")

    # Congelar las capas base
    base_model.trainable = False

    # Añadir capas personalizadas
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(512, activation='relu', kernel_regularizer=l2(0.001))(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    x = Dense(256, activation='relu', kernel_regularizer=l2(0.001))(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    predictions = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=base_model.input, outputs=predictions)
    return model, base_model

# Función para crear y compilar modelos
def build_and_compile_model(model_type, input_shape=(500, 500, 3)):
    if model_type == 'custom_cnn_v2':
        model = create_custom_cnn_v2()
        optimizer = Adam(learning_rate=INIT_LR)
    elif model_type == 'custom_cnn_v3':
        model = create_custom_cnn_v3()
        optimizer = Adam(learning_rate=INIT_LR)
    elif model_type.startswith('tl_'):
        base_model_name = model_type[3:]  # Remover el prefijo 'tl_'
        model, base_model = create_model_with_transfer_learning(base_model_name, input_shape)
        optimizer = Adam(learning_rate=INIT_LR/10)  # Learning rate más bajo para transfer learning
    else:
        raise ValueError(f"Tipo de modelo no reconocido: {model_type}")

    model.compile(
        optimizer=optimizer,
        loss='binary_crossentropy',
        metrics=['accuracy',
                tf.keras.metrics.Precision(name='precision'),
                tf.keras.metrics.Recall(name='recall'),
                tf.keras.metrics.AUC(name='auc')]
    )

    return model

# Lista de modelos a probar
models_to_try = [
    'custom_cnn_v2',
    'custom_cnn_v3',
    'tl_EfficientNetB0',
    'tl_EfficientNetB3',
    'tl_ResNet50',
    'tl_DenseNet121',
    'tl_MobileNetV2',
    'tl_Xception'
]

# Diccionario para almacenar historiales
history_dict = {}
model_performance = {}

# Entrenamiento de modelos
for model_type in models_to_try:
    print(f"\n{'='*60}")
    print(f"Entrenando modelo: {model_type}")
    print(f"{'='*60}")

    # Crear y compilar modelo
    model = build_and_compile_model(model_type)

    # Crear callbacks
    callbacks = create_callbacks(model_type)

    # Entrenar modelo
    history = model.fit(
        train_generator,
        epochs=EPOCHS,
        validation_data=validation_generator,
        callbacks=callbacks,
        verbose=1,
        class_weight=class_weights
    )

    # Guardar historial
    history_dict[model_type] = history

    # Evaluar modelo
    print(f"\nEvaluando modelo {model_type}:")
    evaluation = model.evaluate(validation_generator)

    # Guardar rendimiento
    model_performance[model_type] = {
        'loss': evaluation[0],
        'accuracy': evaluation[1],
        'precision': evaluation[2],
        'recall': evaluation[3],
        'auc': evaluation[4]
    }

    print(f"Pérdida: {evaluation[0]:.4f}, Precisión: {evaluation[1]:.4f}, "
          f"Precisión: {evaluation[2]:.4f}, Recall: {evaluation[3]:.4f}, AUC: {evaluation[4]:.4f}")

    # Guardar modelo
    model.save(f'{model_type}_model.h5')

    # Fine-tuning para modelos de transfer learning
    if model_type.startswith('tl_'):
        print(f"Iniciando fine-tuning para {model_type}")

        # Descongelar las últimas capas del modelo base
        base_model = model.layers[1]  # Asumiendo que el modelo base es la segunda capa
        base_model.trainable = True

        # Congelar las primeras capas y descongelar las últimas
        for layer in base_model.layers[:-20]:
            layer.trainable = False
        for layer in base_model.layers[-20:]:
            layer.trainable = True

        # Recompilar con learning rate más bajo
        model.compile(
            optimizer=Adam(learning_rate=INIT_LR/100),
            loss='binary_crossentropy',
            metrics=['accuracy',
                    tf.keras.metrics.Precision(name='precision'),
                    tf.keras.metrics.Recall(name='recall'),
                    tf.keras.metrics.AUC(name='auc')]
        )

        # Reentrenar con fine-tuning
        history_ft = model.fit(
            train_generator,
            epochs=EPOCHS//2,  # Menos épocas para fine-tuning
            validation_data=validation_generator,
            callbacks=create_callbacks(f"{model_type}_ft"),
            verbose=1,
            class_weight=class_weights
        )

        # Actualizar historial
        for key in history_ft.history.keys():
            history.history[key].extend(history_ft.history[key])

        # Reevaluar después del fine-tuning
        evaluation_ft = model.evaluate(validation_generator)
        model_performance[f"{model_type}_ft"] = {
            'loss': evaluation_ft[0],
            'accuracy': evaluation_ft[1],
            'precision': evaluation_ft[2],
            'recall': evaluation_ft[3],
            'auc': evaluation_ft[4]
        }

        print(f"Después de fine-tuning - Pérdida: {evaluation_ft[0]:.4f}, "
              f"Precisión: {evaluation_ft[1]:.4f}, Precisión: {evaluation_ft[2]:.4f}, "
              f"Recall: {evaluation_ft[3]:.4f}, AUC: {evaluation_ft[4]:.4f}")

        # Guardar modelo fine-tuned
        model.save(f'{model_type}_ft_model.h5')

# Visualización de resultados
plt.figure(figsize=(20, 15))

# 1. Precisión de entrenamiento y validación
plt.subplot(2, 3, 1)
for model_type in history_dict.keys():
    if 'accuracy' in history_dict[model_type].history:
        plt.plot(history_dict[model_type].history['accuracy'],
                 label=f'{model_type} Train', linestyle='--', alpha=0.7)
    if 'val_accuracy' in history_dict[model_type].history:
        plt.plot(history_dict[model_type].history['val_accuracy'],
                 label=f'{model_type} Val', alpha=0.7)
plt.title('Precisión de Entrenamiento y Validación')
plt.ylabel('Precisión')
plt.xlabel('Época')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)

# 2. Pérdida de entrenamiento y validación
plt.subplot(2, 3, 2)
for model_type in history_dict.keys():
    if 'loss' in history_dict[model_type].history:
        plt.plot(history_dict[model_type].history['loss'],
                 label=f'{model_type} Train', linestyle='--', alpha=0.7)
    if 'val_loss' in history_dict[model_type].history:
        plt.plot(history_dict[model_type].history['val_loss'],
                 label=f'{model_type} Val', alpha=0.7)
plt.title('Pérdida de Entrenamiento y Validación')
plt.ylabel('Pérdida')
plt.xlabel('Época')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)

# 3. Métricas de validación comparativas
plt.subplot(2, 3, 3)
metrics = ['accuracy', 'precision', 'recall', 'auc']
width = 0.2
x = np.arange(len(models_to_try))

for i, metric in enumerate(metrics):
    values = [model_performance.get(model, {}).get(metric, 0) for model in models_to_try]
    plt.bar(x + i*width, values, width, label=metric)

plt.title('Métricas de Validación por Modelo')
plt.xlabel('Modelos')
plt.ylabel('Valor')
plt.xticks(x + width*1.5, models_to_try, rotation=45)
plt.legend()
plt.grid(True, axis='y')

# 4. Curvas de aprendizaje de tasa de aprendizaje
plt.subplot(2, 3, 4)
for model_type in history_dict.keys():
    if 'lr' in history_dict[model_type].history:
        plt.plot(history_dict[model_type].history['lr'],
                 label=model_type)
plt.title('Tasa de Aprendizaje Durante el Entrenamiento')
plt.ylabel('Learning Rate')
plt.xlabel('Época')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.yscale('log')

# 5. Matriz de confusión del mejor modelo
best_model_name = max(model_performance, key=lambda x: model_performance[x]['accuracy'])
plt.subplot(2, 3, 5)

# Cargar el mejor modelo
if best_model_name.endswith('_ft'):
    best_model = tf.keras.models.load_model(f'{best_model_name}_model.h5')
else:
    best_model = tf.keras.models.load_model(f'{best_model_name}_model.h5')

# Predecir con el mejor modelo
y_pred = best_model.predict(validation_generator)
y_pred = (y_pred > 0.5).astype(int)
y_true = validation_generator.classes

cm = confusion_matrix(y_true, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=validation_generator.class_indices.keys(),
            yticklabels=validation_generator.class_indices.keys())
plt.title(f'Matriz de Confusión - {best_model_name}')
plt.ylabel('Verdadero')
plt.xlabel('Predicción')

# 6. Resumen de métricas finales
plt.subplot(2, 3, 6)
final_accuracies = [model_performance[model]['accuracy'] for model in models_to_try if model in model_performance]
plt.bar(models_to_try[:len(final_accuracies)], final_accuracies)
plt.title('Precisión Final en Validación por Modelo')
plt.ylabel('Precisión')
plt.xticks(rotation=45)
plt.grid(True)

plt.tight_layout()
plt.savefig('comparacion_modelos_mejorada.png', dpi=300, bbox_inches='tight')
plt.show()

# Identificar el mejor modelo
print(f"\n{'='*60}")
print(f"Mejor modelo: {best_model_name}")
print(f"{'='*60}")
print(f"Métricas del mejor modelo:")
for metric, value in model_performance[best_model_name].items():
    print(f"  {metric}: {value:.4f}")

# Reporte de clasificación del mejor modelo
print('\nReporte de Clasificación:')
print(classification_report(y_true, y_pred,
                            target_names=validation_generator.class_indices.keys()))

# Visualizar algunas predicciones del mejor modelo
def visualize_predictions(model, generator, num_images=10):
    plt.figure(figsize=(15, 10))
    images, labels = next(generator)
    predictions = model.predict(images)

    for i in range(min(num_images, len(images))):
        plt.subplot(2, 5, i+1)
        plt.imshow(images[i])
        true_label = list(generator.class_indices.keys())[int(labels[i])]
        pred_label = list(generator.class_indices.keys())[int(predictions[i] > 0.5)]
        confidence = predictions[i][0] if predictions[i] < 0.5 else 1 - predictions[i][0]

        color = 'green' if true_label == pred_label else 'red'
        plt.title(f'True: {true_label}\nPred: {pred_label}\nConf: {confidence:.2f}', color=color)
        plt.axis('off')

    plt.tight_layout()
    plt.savefig('predicciones_ejemplo_mejoradas.png', dpi=300, bbox_inches='tight')
    plt.show()

visualize_predictions(best_model, validation_generator)

# Resumen de todos los modelos
print(f"\n{'='*80}")
print("RESUMEN COMPARATIVO DE TODOS LOS MODELOS")
print(f"{'='*80}")

performance_df = pd.DataFrame.from_dict(model_performance, orient='index')
print(performance_df.sort_values(by='accuracy', ascending=False))

# Guardar resultados completos
results = {}
for model_type in history_dict.keys():
    results[model_type] = {}
    for key in history_dict[model_type].history.keys():
        results[model_type][key] = history_dict[model_type].history[key]

with open('resultados_entrenamiento_completos.json', 'w') as f:
    json.dump(results, f)

performance_df.to_csv('rendimiento_modelos.csv')
print(f"\nResultados guardados en 'resultados_entrenamiento_completos.json' y 'rendimiento_modelos.csv'")

# Búsqueda de hiperparámetros con Keras Tuner (opcional)
def build_model(hp):
    model = Sequential()

    # Capas convolucionales
    for i in range(hp.Int('num_conv_layers', 2, 5)):
        model.add(Conv2D(
            hp.Int(f'filters_{i}', 32, 256, step=32),
            hp.Choice(f'kernel_size_{i}', [3, 5]),
            activation='relu',
            padding='same'
        ))
        model.add(BatchNormalization())
        if hp.Choice(f'pooling_{i}', ['max', 'avg']) == 'max':
            model.add(MaxPooling2D(2))
        else:
            model.add(layers.AveragePooling2D(2))
        model.add(SpatialDropout2D(hp.Float(f'dropout_conv_{i}', 0.1, 0.4, step=0.1)))

    model.add(GlobalAveragePooling2D())

    # Capas densas
    for i in range(hp.Int('num_dense_layers', 1, 3)):
        model.add(Dense(
            hp.Int(f'units_{i}', 128, 512, step=128),
            activation='relu',
            kernel_regularizer=l2(hp.Float(f'l2_{i}', 1e-5, 1e-2, sampling='log'))
        ))
        model.add(BatchNormalization())
        model.add(Dropout(hp.Float(f'dropout_dense_{i}', 0.3, 0.6, step=0.1)))

    model.add(Dense(1, activation='sigmoid'))

    # Compilar modelo
    model.compile(
        optimizer=Adam(learning_rate=hp.Float('learning_rate', 1e-5, 1e-3, sampling='log')),
        loss='binary_crossentropy',
        metrics=['accuracy', 'precision', 'recall', 'auc']
    )

    return model

# Ejecutar búsqueda de hiperparámetros (descomentar para ejecutar)
"""
tuner = RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=20,
    executions_per_trial=1,
    directory='tuner_results',
    project_name='llama_alpaca_classification'
)

tuner.search(
    train_generator,
    epochs=30,
    validation_data=validation_generator,
    callbacks=[EarlyStopping(patience=5)]
)

# Obtener los mejores hiperparámetros
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
print(f"Mejores hiperparámetros: {best_hps}")

# Construir y entrenar el modelo con los mejores hiperparámetros
best_model = tuner.hypermodel.build(best_hps)
history = best_model.fit(
    train_generator,
    epochs=100,
    validation_data=validation_generator,
    callbacks=create_callbacks('tuned_model'),
    class_weight=class_weights
)
"""
```

```python
import matplotlib.pyplot as plt

def interpolacion_lineal(x1, y1, x2, y2, x):

    if x1 == x2:
        raise ValueError("x1 y x2 no pueden ser iguales (división por cero)")

    y = y1 + (x - x1) * (y2 - y1) / (x2 - x1)
    return y

def graficar_interpolacion(x1, y1, x2, y2, x_interpol, y_interpol):

    plt.figure(figsize=(8, 6))

    plt.plot([x1, x2], [y1, y2], 'bo-', linewidth=2, markersize=8, label='Segmento original')

    plt.plot(x_interpol, y_interpol, 'ro', markersize=8, label='Punto interpolado')


    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Interpolación Lineal Simple')
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.show()

# Ejemplo de uso
if __name__ == "__main__":
    # Definir los dos puntos conocidos
    x1, y1 = 2, 4
    x2, y2 = 8, 10

    # Punto a interpolar
    x = 5

    # Calcular interpolación
    y = interpolacion_lineal(x1, y1, x2, y2, x)

    print(f"Punto 1: ({x1}, {y1})")
    print(f"Punto 2: ({x2}, {y2})")
    print(f"Interpolación en x={x}: y={y:.2f}")

    # Graficar
    graficar_interpolacion(x1, y1, x2, y2, x, y)
```
